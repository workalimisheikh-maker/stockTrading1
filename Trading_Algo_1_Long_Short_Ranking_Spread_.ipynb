```python
date_range = [
    "2023-01-01","2023-02-01","2023-03-01","2023-04-01","2023-05-01","2023-06-01","2023-07-01","2023-08-01",
    "2023-09-01","2023-10-01","2023-11-01","2023-12-01",
    "2024-01-01","2024-02-01","2024-03-01","2024-04-01","2024-05-01","2024-06-01","2024-07-01","2024-08-01",
    "2024-09-01","2024-10-01","2024-11-01","2024-12-01",
    "2025-01-01","2025-02-01","2025-03-01","2025-04-01","2025-05-01","2025-06-01","2025-07-01","2025-08-01"
]


cash = 1000
shares = 0
returns = 1

for x in range(1,len(date_range)-1):
  start_date = date_range[x-1]
  end_date = date_range[x]
  look_forward_date = date_range[x+1]
  price_data = yf.download(Equity_Names,start=start_date,end=end_date)
  price_data = price_data['Close']
  returns_data = price_data.pct_change()[1:]
  factor1 = returns_data.mean()
  ranked_data = factor1.sort_values(ascending=False)
  ranked_data = ranked_data.dropna()
  buy_stock = ranked_data.index[0]
  buy_future_prices = yf.Ticker(buy_stock)
  buy_future_prices = buy_future_prices.history(start=end_date, end=look_forward_date)
  buy_future_prices = buy_future_prices['Close']
  returns = returns * (1+((buy_future_prices[len(buy_future_prices)-1]-buy_future_prices[0])/buy_future_prices[0]))
  sell_stock = ranked_data.index[len(ranked_data)-1]
  sell_future_prices = yf.Ticker(sell_stock)
  sell_future_prices = sell_future_prices.history(start=end_date, end=look_forward_date)
  sell_future_prices = sell_future_prices['Close']
  returns = returns * (1+((sell_future_prices[0]-sell_future_prices[len(sell_future_prices)-1])/sell_future_prices[0]))

  print("\n Date:", date_range[x])
  print("Stock to Buy:",buy_stock)
  print("Returns (Reason):",ranked_data[0])
  print("Current Price:",buy_future_prices[0])
  print("Next Month Price:",buy_future_prices[len(buy_future_prices)-1])
  print("Stock to Short:",sell_stock)
  print("Returns (Reason):",ranked_data[len(ranked_data)-1])
  print("Current Price:",sell_future_prices[0])
  print("Next Month Price:",sell_future_prices[len(buy_future_prices)-1])
  print("Return for that month:",((sell_future_prices[0]-sell_future_prices[len(sell_future_prices)-1])/sell_future_prices[0]))
  print("Cumulative Returns:",returns,"\n")



```
